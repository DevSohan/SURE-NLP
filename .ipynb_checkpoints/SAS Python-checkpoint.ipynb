```python
# Bestfit Plane for Waterline
# https://scikit-spatial.readthedocs.io/en/stable/gallery/fitting/plot_plane.html
from skspatial.objects import Plane, Points
from skspatial.plotting import plot_3d
```

```python
points = Points([[96.4868, 106.1078, 8.7028],
[96.5988, 108.3448, 8.7642],
[96.3928, 104.7255, 8.7099],
[97.5255, 102.4448, 8.7195],
[99.0977, 106.2245, 8.7222],
[98.6674, 102.8796, 8.7347]])
```

```python
plane = Plane.best_fit(points)
plane
```

```python
plot_3d(
    points.plotter(c='k', s=50),
    plane.plotter(alpha=0.5)
)
```
